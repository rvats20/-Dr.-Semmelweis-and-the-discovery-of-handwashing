```python
# Import libraries
import matplotlib.pyplot as plt
from numpy import asarray
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
# Custom data
dataset = asarray([[30, 0.001],
                   [18,  0.05],
                   [88,  0.005],
                   [50,  0.001],
                   [4,   0.1],
                   [10,  0.001],
                   [38,  0.01],
                   [40,  0.05],
                   [44,  0.47],
                   [114, 0.01]])
```

```python
# Line plot 2D data function
def plot_2D_data(data, title):
  x = range(len(data))
  plt.rcParams["figure.figsize"] = (10, 5)
  plt.plot(x, data.T[0], marker='o', label='Attribute 1')
  plt.plot(x, data.T[1], marker='o', label='Attribute 2')
  plt.legend()
  plt.title('Plot ' + title, size=16)
  plt.xlabel('x')
  plt.ylabel('value')
  plt.grid()
  plt.show()
```

```python
# Line plot regular data
plot_2D_data(dataset, "Regular Dataset")
```

```python
# Blox plot 2D data function
def boxplot_2D_data(data, title):
  plt.rcParams["figure.figsize"] = (10,5)
  plt.boxplot(data, labels=['Attribute 1', 'Attribute 2'], showmeans=True)
  plt.title('Boxplot ' + title, size=16)
  plt.grid()
  plt.show()
```

```python
# Box plot regular data
boxplot_2D_data(dataset, "Regular Dataset")
```

## <span>1. Normalization</span>

```python
# Define min-max scaler
scaler_1 = MinMaxScaler()

# Transform data
scaled_1 = scaler_1.fit_transform(dataset)
```

```python
# Line plot normalized data
plot_2D_data(scaled_1, "Normalized Dataset")
```

```python
# Box plot regular data
boxplot_2D_data(scaled_1, "Normalized Dataset")
```

## <span>2. Standardization</span>

```python
# Define standard scaler
scaler_2 = StandardScaler()

# Transform data
scaled_2 = scaler_2.fit_transform(dataset)
```

```python
# Line plot standardized data
plot_2D_data(scaled_2, "Standardized Dataset")
```

```python
# Box plot regular data
boxplot_2D_data(scaled_2, "Standardized Dataset")
```

## <span>3. Robust Scaling</span>

```python
# Define robust scaler
scaler_3 = RobustScaler()

# Transform data
scaled_3 = scaler_3.fit_transform(dataset)
```

```python
# Line plot standardized data
plot_2D_data(scaled_3, "Robusted Dataset")
```

```python
# Box plot regular data
boxplot_2D_data(scaled_3, "Robusted Dataset")
```
